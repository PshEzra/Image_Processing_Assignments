```python
# Import packages
from skimage import io, color
from skimage.morphology import disk, square, diamond, dilation, erosion, closing, opening, black_tophat, white_tophat
import matplotlib.pyplot as plt
```

# Nomor 1

```python
ori_img = color.rgb2gray(io.imread("shape.png", pilmode = 'RGB'))
plt.figure(figsize =(10 , 5))
plt.imshow(ori_img, cmap='gray')
plt.show()
```

### 1.a

```python
closed_img = closing(ori_img, disk(65))
erroded_img = erosion(closed_img, disk(70))
dil_img = dilation(erroded_img, disk(70))

plt.figure(figsize =(10 , 5))
plt.imshow(dil_img, cmap='gray')
plt.show()
```

### 1.b

Tiga transformasi yang dilakukan adalah closing, erroding, dan dilating

> Closing dilakukan untuk menutup bagian citra yang tampak terbuka seperti di bawah setiap segitiga dan di ujung kotak. Dengan dilakukan closing, bagian citra tersebut menjadi tertutup dan membuat transformasi ke citra yang diinginkan menjadi lebih mudah.

> Erroding dilakukan untuk mengecilkan luas setiap elemen citra. Setiap elemen dalam citra harapan terlihat lebih kecil dibandingkan citra asli. Ini dilakukan karena errosion mengurangi luas dari setiap elemen sehingga. Erroding juga dilakukan agar proses dilation dapat menghasilkan citra harapan melalui transformasinya.

> Dilating dilakukan untuk memperluas citra yang sudah di-errode. Dengan memperluas setiap elemen dengan strel disk, citra yang di dilasi memiliki hasil yang bulat dan tidak memiliki ujung tajam di setiap elemennya.

# Nomor 2

```python
ori_img_2 = color.rgb2gray(io.imread("AI.png", pilmode = 'RGB'))
plt.figure(figsize =(10 , 5))
plt.imshow(ori_img_2, cmap='gray')
plt.show()
```

## 2.a

```python
er_img = erosion(ori_img_2, disk(5))

plt.figure(figsize =(10 , 5))
plt.imshow(er_img, cmap='gray')
plt.show()
```

```python
final_img2 = er_img - ori_img_2

plt.figure(figsize =(10 , 5))
plt.imshow(final_img2, cmap='gray')
plt.show()
```

### 2.b

Citra harapan merupakan *outline* dari citra asli. Untuk mendapatkan ini, kita perlu menebalkan luas dari citra asli tersebut. Kemudian, kita mengurangi citra yang ditebalkan dengan citra asli.

Untuk menebalkan citra kita menggunakan proses erosi. Erosi digunakan karena elemen utama dari citra memiliki warna yang lebih gelap dari background. Hasil penebalan adalah gambar kedua di atas.

Untuk proses akhir, kita dapat dengan mudah mengurangi citra yang ditebalkan dengan citra original sehingga didapatkan gambar outline.

# Nomor 3

```python
ori_img_3 = color.rgb2gray(io.imread("road_sign1.jpg", pilmode = 'RGB'))
plt.figure(figsize =(10 , 5))
plt.imshow(ori_img_3, cmap='gray')
plt.show()
```

```python
ori_img_3b = color.rgb2gray(io.imread("road_sign2.jpg", pilmode = 'RGB'))
plt.figure(figsize =(10 , 5))
plt.imshow(ori_img_3b, cmap='gray')
plt.show()
```

### 3.a.

```python
text_img = black_tophat(ori_img_3, disk(50))

plt.figure(figsize =(10 , 5))
plt.imshow(text_img, cmap='gray')
plt.show()
```

```python
text_imgb = white_tophat(ori_img_3b, disk(50))

plt.figure(figsize =(10 , 5))
plt.imshow(text_imgb, cmap='gray')
plt.show()
```

### 3.b.

Kedua citra didapatkan menggunakan tophat filtering. Pada citra pertama dilakukan black tophat filter karena teks pada citra memiliki warna hitam. Pada citra kedua dilakukan white tophat filter karena teks pada citra memiliki warna putih.

Hasil yang didapatkan cukup baik dalam menonjolkan fitur teks pada citra namun dengan melakukan filter tophat, beberapa fitur yang tidak diinginkan juga lolos filter seperti gedung pada citra pertama.
